```python
import pandas as pd
import os
import re
```

```python
def add_record(df, path, text, label):
    with open(os.path.join(path, text), encoding='utf-8') as file:
        review = file.read()
    voto = re.search(r'_(\d+)\.txt', text).group(1)
    new_record= {df.columns[0] : review, 
                df.columns[1] : label, 
                df.columns[2] : voto}
    
    df.loc[len(df)] = new_record
```

```python
reviews_data = pd.DataFrame(columns=['review', 'label', 'voto'])

for path, folder, files in os.walk('Dati/aclImdb/train'):
    if not folder and re.search(r'neg$', path, re.IGNORECASE):
        for file in files:
            add_record(reviews_data, path, file, 'Negativa')
    elif not folder and re.search(r'pos$', path, re.IGNORECASE):
        for file in files:
            add_record(reviews_data, path, file, 'Positiva')
    elif not folder and re.search(r'unsup$', path, re.IGNORECASE):
        for file in files:
            add_record(reviews_data, path, file, 'NULL')        
```

```python
display(reviews_data)
reviews_data.to_csv('Dati/train.csv', index=False)
```
